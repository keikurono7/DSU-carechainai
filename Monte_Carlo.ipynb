```python
import matplotlib.pyplot as plt
import random
import numpy as np

rlist1=[]
rlist2=[]

while len(rlist1)<100 or len(rlist2)<100:
    if len(rlist1)<100:
        r=random.randint(1,100)
        if r not in rlist1:
            rlist1.append(r)
    if len(rlist2)<100:
        r=random.randint(1,100)
        if r not in rlist2:
            rlist2.append(r)
    
x=np.array(rlist1)
y=np.array(rlist2)

plt.title('Monte carlo')
plt.scatter(x,y)
circle = plt.Circle((50, 50), 50, color='r', fill=False)
plt.gca().add_artist(circle)
d = np.sqrt((x - 50) ** 2 + (y - 50) ** 2)  # distance from each point to the center
n_inside = np.sum(d <= 50)
n_outside = np.sum(d > 50)

print(f"Number of points inside the circle: {n_inside}")
print(f"Number of points outside the circle: {n_outside}")

print(4*(n_outside/n_inside))
plt.show()
```
